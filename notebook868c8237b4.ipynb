```python
import numpy as np
import pandas as pd



my_data= pd.read_csv('/kaggle/input/data-set-1/cadastre.csv')
```

```python
my_data.isna().sum()
```

```python
my_data = my_data [columns =  
    '#_X', 'Current_Yield1'
    '#_Y', 'Past_Year_Yeild2']
```

```python
my_data.columns= ['Current_Yield', 'Past_Yield', 'State', 'SA4']

my_data.head()
my_data.columns
```

```python
my_data['row_number']= range (1, len(my_data) +1)

my_data.head()
```

```python
my_data ['Past_Yield']= my_data['Past_Yield']. apply(lambda x: -x if x<0 else x)

my_data.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt



my_data_subset= my_data.iloc[0:50] 

plt.figure(figsize=(8, 5))
plt.bar(my_data_subset['Current_Yield'], my_data_subset['Past_Yield'])
plt.title('Rows 1 to 50')
plt.xlabel('Current yield')
plt.ylabel('Past Yield')
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns

sns.barplot(x='Current_Yield', y='Past_Yield', data=my_data)
plt.show()
```
